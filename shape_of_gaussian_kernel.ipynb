```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
```

Instantiating models

```python
svc_linear = SVC(kernel='linear')
svc_poly = SVC(kernel='poly', degree=3)
svc_rbf = SVC(kernel='rbf')
```

Creating dataset

```python
# Generate data
np.random.seed(0)
X = np.random.randn(100, 2) # 100 points in 2D
y = np.logical_xor(X[:, 0] > 0, X[:, 1] > 0) # XOR function
```

Plot Data

```python
# Plot data
plt.scatter(X[:, 0], X[:, 1], c = y)
```

Fitting models

```python
# Fit models
svc_linear.fit(X, y)
svc_poly.fit(X, y)
svc_rbf.fit(X, y)
```

Testing accuracy

```python
# Test accuracy
print('Linear accuracy:', svc_linear.score(X, y))
print('Poly accuracy:', svc_poly.score(X, y))
print('RBF accuracy:', svc_rbf.score(X, y))
```

Plotting decision boundaries

```python
# Plot decision boundaries
plt.figure()
plt.scatter(X[:, 0], X[:, 1], c = y)
h = .02
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = svc_linear.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.8)
plt.title('Linear kernel')
```

Now, the main kernel to test

```python
plt.figure()
plt.scatter(X[:, 0], X[:, 1], c = y)
Z = svc_rbf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.8)
plt.title('RBF kernel')
```

Let's test sigmoid as well

```python
# test sigmoid
svc_sigmoid = SVC(kernel='sigmoid')
svc_sigmoid.fit(X, y)
print('Sigmoid accuracy:', svc_sigmoid.score(X, y))
plt.figure()
plt.scatter(X[:, 0], X[:, 1], c = y)
Z = svc_sigmoid.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.8)
plt.title('Sigmoid kernel')
```
